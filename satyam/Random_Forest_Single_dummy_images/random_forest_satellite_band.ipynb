```python
from sklearn.ensemble import RandomForestRegressor
import numpy as np

import matplotlib.pyplot as plt; plt.rcdefaults()
```

```python
import PIL
```

```python
# Create dataset
# Satellite 2
Sa2 = PIL.Image.open("dummy/010214A1.png").convert("L")
Sa2arr = np.array(Sa2).flatten()
Sa2arr.shape
```

```python
# Satellite 3
Sa3 = PIL.Image.open("dummy/010214A2.png").convert("L")
Sa3arr = np.array(Sa3).flatten()
Sa3arr.shape
```

```python
# Satellite 4
Sa4 = PIL.Image.open("dummy/010214A3.png").convert("L")
Sa4arr = np.array(Sa4).flatten()
Sa4arr.shape
```

```python
# Satellite 6
Sa6 = PIL.Image.open("dummy/010214A4.png").convert("L")
Sa6arr = np.array(Sa6).flatten()
Sa6arr.shape
```

```python
# Radar Output
Radar = PIL.Image.open("dummy/010214B1.png").convert("L")
Radararr = np.array(Radar).flatten()
Radararr.shape
```

```python
Y = Radararr
```

```python
# Create rows for the dataset
X = [];
for i in range(65536):
    x = [];
    x.append(Sa2arr[i])
    x.append(Sa3arr[i])
    x.append(Sa4arr[i])
    x.append(Sa6arr[i])
    X.append(x)
```

```python
X_numpy = np.asarray(X)
X_numpy
```

```python
X_numpy.shape
```

```python
names = ["Band 2", "Band 3", "Band 4", "Band 6"]
X_names = np.asarray(names)
```

# Fit the model

```python
rf = RandomForestRegressor()
rf.fit(X_numpy, Y)
```

```python
for feat, score in zip(X_names, rf.feature_importances_):
    print('{}       {}'.format(feat, score))
```

```python
plt.bar(names,rf.feature_importances_, color='g')
plt.xlabel('Feature')
plt.ylabel('Score')
```

```python
plt.title('Feature Importance')
plt.show()
```
