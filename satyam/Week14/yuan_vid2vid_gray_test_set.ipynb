```python
def rmse(imageA, imageB):
    err = np.sum((imageA.astype("float") - imageB.astype("float")) ** 2)
    err /= float(imageA.shape[0] * imageA.shape[1])
    err = math.sqrt(err)
    return err
```

```python
ssim_list = []
rmse_list = []

indexOf_least_rsme = 0;
least_rsme = 99999;
indexOf_largest_rsme = 0;
largest_rsme = 0;

indexOf_least_ssim = 0;
least_ssim = 9;
indexOf_largest_ssim = 0;
largest_ssim = -9;

for i in range(len(original)):
    print(predicted[i])
    print(original[i])
    fig=plt.figure()
    pred=np.asarray(Image.open(predicted[i]))
    real=np.asarray(Image.open(original[i]))
    plt.xlabel('Predicted', fontsize=20)
    plt.imshow(pred, cmap="gray")
    plt.show()
    plt.xlabel('Actual', fontsize=20)
    plt.imshow(real, cmap="gray")
    plt.show()
    rsme_ = rmse(pred, real)
    ssim_ = ssim(pred, real)
    print("RMSE: ", rsme_)
    rmse_list.append(rsme_)
    print("SSIM: ", ssim_)
    ssim_list.append(ssim_)
    
    if(rsme_ < least_rsme):
        least_rsme = rsme_
        indexOf_least_rsme = i
    
    if(ssim_ < least_ssim):
        least_ssim = ssim_
        indexOf_least_ssim = i
        
    if(rsme_ > largest_rsme):
        largest_rsme = rsme_
        indexOf_largest_rsme = i
    
    if(ssim_ > largest_ssim):
        largest_ssim = ssim_
        indexOf_largest_ssim = i
        
        
    
    print("\n")
    print("=========================================")
    
```
